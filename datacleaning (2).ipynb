```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()
df
```

```python
#display columns with missing values
missing_columns= missing_values[missing_values > 0]
print(missing_columns)
```

```python
# Handle missing values
# For simplicity, let's fill missing numeric values with the median and categorical with the mode
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        # Using loc to avoid chained assignment
        df.loc[:, column] = df[column].fillna(df[column].median())
    else:
        # Using loc to avoid chained assignment for non-numeric columns
        df.loc[:, column] = df[column].fillna(df[column].mode()[0])

# Verify that there are no more missing values
missing_values_after = df.isnull().sum().sum()
print("Total missing values after handling:", missing_values_after)
```

```python
df = df.dropna()
df
```

```python
print("Data cleaning, validation, and preprocessing completed.")
print("\
Dataset shape:", df.shape)
print("\
Column types:")
print(df.dtypes)

# Display the first few rows of the preprocessed dataset
print("\
First few rows of the preprocessed dataset:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
plt.style.use('default')

# Visualize distributions of key numerical features
numerical_features = ['Glacier_Mean_Summer_Discharge_GtPerYr', 'Glacier_Mean_Summer_Discharge_Error_GtPerYr', 'Glacier_Mean_Autumn_Discharge_GtPerYr','Cumulative_FWF_m3_MAR','AnnualMean_FWF_m3_MAR','Mean_Winter_FWF_m3_MAR','Mean_Spring_FWF_m3_MAR','Glacier_Point_Latitude','Glacier_Point_Longitude']

fig, axes = plt.subplots(3, 3, figsize=(20, 20))
axes = axes.flatten()

for i, feature in enumerate(numerical_features):
    sns.histplot(df[feature].dropna(), kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()

# Analyze relationships between features
correlation_matrix = df[numerical_features].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
print("EDA completed. Plots displayed in the notebook.")
```

```python
# Updated EDA Exploration Script
import matplotlib.pyplot as plt
import seaborn as sns

# Define the new numerical columns based on the provided column names
numerical_features = [
    'Glacier_Mean_Summer_Discharge_GtPerYr', 
    'Glacier_Mean_Summer_Discharge_Error_GtPerYr', 
    'Glacier_Mean_Autumn_Discharge_GtPerYr',
    'Cumulative_FWF_m3_MAR', 
    'AnnualMean_FWF_m3_MAR', 
    'Mean_Winter_FWF_m3_MAR', 
    'Mean_Spring_FWF_m3_MAR'
]

# Define new categorical columns for visualizations (use placeholders or relevant columns as needed)
categorical_features = ['Glacier_Point_Latitude', 'Glacier_Point_Longitude']

# 1. Pairplot to visualize relationships between numerical features
sns.pairplot(df[numerical_features])
plt.suptitle('Pairplot of Numerical Glacier Features', y=1.02)
plt.show()

# 2. Boxplot to identify outliers in numerical features
plt.figure(figsize=(15, 10))
sns.boxplot(data=df[numerical_features])
plt.title('Boxplot of Numerical Glacier Features')
plt.xticks(rotation=45)
plt.show()

# 3. Countplot for the updated categorical features
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=df[feature])
    plt.title(f'Countplot of {feature}')
    plt.show()

print("Additional EDA exploration completed.")
```
